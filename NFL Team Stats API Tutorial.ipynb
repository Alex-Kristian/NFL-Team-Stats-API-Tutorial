# NFL Team Stats API Tutorial

# Import Requests Library

```python
import requests
```

# Get Team Receiving Data

```python
url = "https://nfl-team-stats.p.rapidapi.com/v1/nfl-stats/teams/receiving-stats/offense/2022"

headers = {
    "X-RapidAPI-Key": "ca4e0f7580mshf9be837b2c3db25p1d82d9jsn6f5cd82d4d53",
    "X-RapidAPI-Host": "nfl-team-stats.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers)

x = list(list(response.json().values())[0].values())[0]

for teams in x:
    for stats in teams:
        print(stats,":", teams.get(stats))
    print("\n")


```
